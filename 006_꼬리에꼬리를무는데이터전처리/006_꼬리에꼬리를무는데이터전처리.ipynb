### order_status 컬럼 생성하기

```python
# order_status 열 추가
transactions['order_status'] = transactions['total_amt'].apply(lambda x: 'Canceled' if x < 0 else 'Ordered')
transactions.head()
```

```python
# sampling data
transactions[transactions['transaction_id'] == 80712190438].sort_values(by='tran_date')
```

```python
transactions.groupby('order_status').size()
```

```python
order_minus_cancel = 20876-2177
order_minus_cancel
```

### confirmed_yn 컬럼 생성하기

```python
# step1. order_status의 Canceled 값만 필터링 하기
canceled_df = transactions.loc[transactions['order_status'] == 'Canceled']
canceled_df.head()
```

```python
# step2. transaction_id만 select한 후 confirmed_yn 컬럼 생성하여 해당 값에 0 부여하기
canceled_df = canceled_df.loc[:,['transaction_id']].assign(confirmed_yn = 0)
canceled_df.head()
```

```python
canceled_df.info()
```

```python
# stpe3. 중복값 제외하기
canceled_df = canceled_df.drop_duplicates()
canceled_df.info()
```

```python
# canceled_df 정보 join
transactions = transactions.merge(canceled_df, on='transaction_id', how='left')
transactions.head()
```

```python
# sampling data
transactions[transactions['transaction_id'] == 80712190438].sort_values(by='tran_date')
```

```python
# confirmed_yn 열을 기준으로 그룹화하고 행의 개수 계산
transactions.groupby('confirmed_yn').size()
```

```python
# confirmed_yn 열을 기준으로 그룹화하고 NA값을 대체한 후 행의 개수 계산
transactions['confirmed_yn'].fillna(1).groupby(transactions['confirmed_yn'].fillna(1)).size()
```

```python
# confirmed_yn 열의 NA 값을 1로 대체
transactions['confirmed_yn'] = transactions['confirmed_yn'].fillna(1)
transactions.head(20)
```

```python
# confirmed_yn 값으 기준으로 데이터 집계
confirmed_yn_result = transactions.groupby('confirmed_yn').size()
confirmed_yn_result
```

```python
order_minus_cancel == confirmed_yn_result[1.0]
```

### 주문은 없고, 취소가 발생한 건이 있는지 확인

```python
od_ca_result = transactions2.groupby('transaction_id').agg(
    od_n=pd.NamedAgg(column='order_status', aggfunc=lambda x: (x == 'Ordered').sum()),
    ca_n=pd.NamedAgg(column='order_status', aggfunc=lambda x: (x == 'Canceled').sum())
).reset_index()
od_ca_result.head()
```

```python
# 'od_n'이 0인 행만 필터링
omit_df = od_ca_result[od_ca_result['od_n'] == 0]
omit_df
```

```python
# sampling data
transactions2[transactions2['transaction_id'] == 8868056339].sort_values(by='tran_date')
```
